```python
import requests
from selenium import webdriver
import time
import pandas as pd
from pandas import DataFrame,Series
from bs4 import BeautifulSoup as bs4
import re
```

```python
def get_table(url):
    
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.132 Safari/537.36'}
    options = webdriver.ChromeOptions()
    options.add_argument('--headless')#不弹窗
    options.add_argument("–proxy-server=http://14.20.235.95:9797")# 一定要注意，等号两边不能有空格
    browser = webdriver.Chrome(options = options )

    browser.get(url)
    js="var q=document.documentElement.scrollTop=100000"  #注意：此必须处加入向下滑动操作，否则提取不到源代码
    browser.execute_script(js)  
    time.sleep(3)

    data = browser.page_source
    # print(data)
    resou = bs4(data, 'lxml')
#     print(resou.prettify())
    title = resou.title.text
    title = re.sub('/','-',title)  #注意：文件命名不能含有'/'等特殊符号，所以需要提前替换
#     len(resou.tbody.find_all(style = "height: 13.5pt;"))
    num_h = len(resou.tbody.find_all('tr')) #计算表格的行数
    table = resou.tbody.find_all('td')
    num_total =  len(table) # 表格中的总格子数
    num_l = num_total//num_h #计算表格的列数
    lies = []
    for i in range(num_l):
        lie = table[i::num_l]
        lies.append(lie)   #   lies[1][2].text # lies[列数][行数],此时的lies是各列的源代码


    for i in range(len(lies)):
        for j in range(len(lies[i])):
            lies[i][j] = lies[i][j].text # 此时的lies就是各列的文本了
            
    df = DataFrame() #建立一个dataframe空表格
    for i in range(len(lies)):
        df[i] = lies[i]  # 按列将lies的各列加到空表格中
#     return df        # 返回表格，大功告成！！
#     path = 'C:/Users/Bill/Desktop/qichexiehui/'
    df.to_csv('C:\\Users\\Bill\\Desktop\\汽车工业协会表格爬取\\亚洲\\'+title+'.csv') #保存
```

```python
   
def get_hrefs(url_list):

    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.132 Safari/537.36'}
    options = webdriver.ChromeOptions()
    options.add_argument('--headless')#不弹窗
    options.add_argument("–proxy-server=http://14.20.235.95:9797")# 一定要注意，等号两边不能有空格
    browser = webdriver.Chrome(options = options )

    browser.get(url_list)
    js="var q=document.documentElement.scrollTop=100000"  #注意：此必须处加入向下滑动操作，否则提取不到源代码
    browser.execute_script(js)  
    time.sleep(3)

    data = browser.page_source
    browser.close()
    data
    
    hrefs = re.findall('<a href="con_(.*?)" target="_blank">.*?<span class="point">',data,re.S)

    address = []
    for i in range(len(hrefs)):
        hrefs[i] = 'http://www.caam.org.cn/chn/4/cate_41/con_'+hrefs[i]
        address.append(hrefs[i])
    return address
```

```python
# cates = ['50':'欧洲', '41';'亚洲','56':'大洋','59':'北美','62':'南美','66':'非洲']

page = 88
queshi = []
while page < 304:
    url_list = 'http://www.caam.org.cn/chn/4/cate_41/list_'+str(page)+'.html'
    for url in get_hrefs(url_list):
        try :
            get_table(url)
        except:
            print(url)
            queshi.append(url)
    page += 1
```
